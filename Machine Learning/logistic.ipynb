```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('breast_cancer.csv')
```

```python
df.head(10)
```

```python
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
df.head()
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
y.replace('M', 1, inplace=True)
y.replace('B', 0, inplace=True)
y.head()
```

```python
X.head()
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
def sigmoid_func(z):
    return (1 / (1 + np.exp(-z)))
```

```python
def cost_func(X, y, theta):
    hyp = sigmoid_func(np.dot(X, theta))
    return (-(y.T * np.log(hyp) + (1 - y).T * np.log(1 - hyp))).mean()
```

```python
def gradient_descent(X, y, epochs, alpha=0.01):
    theta = np.zeros(X.shape[1])
    cost_history = []
    for i in range(epochs):
        y_pred= sigmoid_func(np.dot(X, theta))
        gradient = (alpha/X.shape[0]) * ((y_pred-y).dot(X)) 
        theta -= gradient
        cost = cost_func(X, y, theta)
        cost_history.append(cost)
    return theta,cost_history
```

```python
theta,cost_history = gradient_descent(X_train, y_train, 1000)
print(len(theta))
cost_history
```

```python
y_pred = sigmoid_func(X_test.dot(theta))
y_label= (y_pred>0.5).astype(int)
```

```python
from sklearn import metrics
print("The Accuracy score is : ", (metrics.accuracy_score(y_test,y_label)*100))
```

```python
plt.plot(np.arange(1000),cost_history,random_state=1)
plt.xlabel("No. of iterations")
plt.ylabel("Cost")
plt.legend()
plt.show()
```
